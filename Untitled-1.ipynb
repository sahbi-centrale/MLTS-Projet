# IMPORT MODULE

```python
import pandas as pd
import requests
from io import StringIO
import numpy as np 
import matplotlib.pyplot as plt
import adtk
from  adtk.data import validate_series
import matplotlib.style
%matplotlib inline
```

---

# IMPORT DATA SET

```python
# URL of the CSV file
url = 'https://raw.githubusercontent.com/numenta/NAB/master/data/realKnownCause/ambient_temperature_system_failure.csv'

# Use requests to get the content of the CSV file without verifying SSL certificate
# Note: Disabling SSL verification is not recommended for production code
response = requests.get(url, verify=False)

# Check if the request was successful
if response.status_code == 200:
    # Use StringIO to convert the text content into a file-like object
    data = StringIO(response.text)
    # Read the data into a pandas DataFrame
    temperature_data = pd.read_csv(data)
    # Output the first few rows of the DataFrame
    print(temperature_data.head())
else:
    print(f"Failed to download data, status code: {response.status_code}")
```

---

#  COMPARAISON BETWEEN THE TimeEval AND THE ADTK

# ADTK

```python
# Assuming 'temperature_data' is a DataFrame and it has a column 'date' with date-time information
temperature_data['date'] = pd.to_datetime(temperature_data['timestamp'])
temperature_data.set_index('date', inplace=True)
temperature_data = temperature_data.drop('timestamp', axis = 1)

# Now you can use 'temperature_data' with 'validate_series'
temperature_data = validate_series(temperature_data)
```

# PLOTING THE TIME SERIES PLOT

```python
plt.figure(figsize=(10, 6))
plt.plot(temperature_data.index, temperature_data['value'])  # Replace 'your_value_column' with your data column name
plt.title('Time Series Plot')
plt.xlabel('Date')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

# HISTOGRAM 

```python
plt.hist(temperature_data , bins = 20)
```

The very first way to identify anomalous points that is presented here is to ignore the time dependencies and to look at the points the histogram of the values and do a threshold on the values that way.

# adtk.detector.ThresholdAD

```python
from adtk.detector import ThresholdAD

threshold_ad = ThresholdAD(high = 80 , low = 60)

anomalies = threshold_ad.detect(temperature_data)

temperature_data['anomalie'] = anomalies['value'].values
```

```python
plt.figure(figsize=(10, 6))
plt.plot(temperature_data.index, temperature_data['value'], label='Time Series Data')

# Overlay the anomalies
# This will plot red points at the locations of the anomalies
anomalies = temperature_data[temperature_data['anomalie']]
plt.scatter(anomalies.index, anomalies['value'], color='red', label='Anomalies')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series with Anomalies')
plt.legend()

# Show the plot
plt.show()
  # Filter rows where anomalies are True
```

# adtk.detector.QuantileAD

# Fitting on the basis of historical datas 

```python
from adtk.detector import QuantileAD

temperature_data = temperature_data.drop('anomalie' , axis = 1)

quantile_ad = QuantileAD(high = 0.99 , low = 0.01)

anomalies_1 = quantile_ad.fit_detect(temperature_data)

temperature_data['anomalie'] = anomalies_1['value'].values
```

```python
plt.figure(figsize=(10, 6))
plt.plot(temperature_data.index, temperature_data['value'], label='Time Series Data')

# Overlay the anomalies
# This will plot red points at the locations of the anomalies
anomalies_2 = temperature_data[temperature_data['anomalie']]
plt.scatter(anomalies.index, anomalies['value'], color='red', label='Anomalies')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series with Anomalies')
plt.legend()

# Show the plot
plt.show()
  # Filter rows where anomalies are True
```
